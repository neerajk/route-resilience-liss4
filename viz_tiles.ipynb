```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import random

# Helper function for contrast stretching (essential for viewing raw satellite DNs)
def stretch(band, p_lower=2, p_upper=98):
    if not band.any(): return np.zeros_like(band)
    b_min, b_max = np.percentile(band[band > 0], (p_lower, p_upper))
    return np.clip((band - b_min) / (b_max - b_min), 0, 1)
```
